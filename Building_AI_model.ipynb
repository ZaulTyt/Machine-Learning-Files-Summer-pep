```python
import requests
url="https://codeunnati.edunetfoundation.com/faqs"

data=requests. get(url)
```

```python
with open("codeunnati.pdf", "wb") as f:
    f.write(data.content)

text_splitter = CharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=100
)

from langchain_core import documents

with open("codeunnati.pdf", "rb") as f:
    loader = UnstructuredFileIOLoader(f)
    documents = loader.load()

texts = text_splitter.split_documents(documents)
```

```python
texts[0]
```

```python
embedding=HuggingFaceEmbeddings()
```

```python
presist_directory="vector_db"
```

```python
chroma=Chroma.from_documents(documents=texts,embedding=embedding,persist_directory=presist_directory)
```

```python
retriever=chroma.as_retriever()
```

```python
llm=ChatGroq(model="llama-3.1-8b-instant",temperature=0)
```

```python
qa_chain=RetrievalQA.from_chain_type(llm=llm,chain_type="stuff",retriever=retriever)
```

```python
query="What is code unnati program"
response=qa_chain.invoke({"query":query})
print(response["result"])
```

```python
import gradio as gr

def get_response(question):
  try:
    response=qa_chain.invoke({"query":question})
    return response["result"]
  except Exception as e:
    return str(e)


demo=gr.Interface(
    fn=get_response,
    inputs=gr.Text(
        lines=2,
        placeholder="Enter your query related to codeunnati program"

    ),
    outputs=gr.Text(
        lines=10,
        label="Response"

    ),
     title="Code unnatio Chat Bot",
     description="Ask the query related to the Codeunnati program"

    )
demo.launch(debug=True)
```
